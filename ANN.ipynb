```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Churn_Modelling.csv')
X = data.iloc[:, 3:13].values
y = data.iloc[:, 13].values
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le_1 = LabelEncoder()
X[:,1] = le_1.fit_transform(X[:,1]) # Generating numerical labels for countries

le_2 = LabelEncoder()
X[:,2] = le_2.fit_transform(X[:,2]) # Generating numerical labels for gender

one_hot = OneHotEncoder(categorical_features=[1])
X = one_hot.fit_transform(X).toarray() # dummy variables for countries
X = X[:, 1:] # droping one country dummy variable to avoid dummy variable trap
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.25)
```

```python
# Feature Scaling
# to bring all features in same range values
# this way one feature will not dominate another feature
# important step in NN and DL

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Building ANN with keras

from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
# Adding hidden layer
# uniform function initialize weights and weights are small close to zero.
# 11 input layer nodes for 11 independent features
# 6 hidden layer nodes(av. of i/p and o/p layer nodes)(no fixed rule)
model.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim = 11)) 

# Adding another hidden layer to implement deep NN
model.add(Dense(units=6, kernel_initializer='uniform', activation='relu')) # rectifier activation function

# Adding output layer
# sigmoid activation function to determine the probabilities which customers will leave the bank.
# output layer node is 1(either leave bank or not)
# for more than 2 categories in output variable, softmax activation is used.
model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid')) 

# Compile ANN by stochastic gradient descent
# optimizer will update weights
# loss function is optimized to find optimal weights
# loss= categorical_crossentropy from more than 2 outcomes in output
# accuracy criterion to evaluaue and improve the model.
model.compile(optimizer= 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting model
# after a batch, weights are updated.
model.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
# probabilities of customers leaving the bank.(due to sigmoid function)
# these are not 0 or 1 values.
y_pred = model.predict(X_test)
```

```python
y_pred = (y_pred > 0.5) # converting probabilities into 0 and 1.
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```

86% accuracy on test set
